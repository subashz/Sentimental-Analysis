```python
contractions_re=re.compile('(%s)' % '|'.join(contractions.keys()))
def expand_contractions(text,contractions=contractions):
    def replace(match):
        return contractions[match.group(0)]
    return contractions_re.sub(replace, text)
# Expanding Contractions in the reviews
df['Review']=df['Review'].apply(lambda x:expand_contractions(x))
df.head()
```

```python
# removing punctuation

print(string.punctuation)

def remove_punc(txt):
    txt_nopunct = "".join([x for x in txt if x not in string.punctuation])
    return txt_nopunct

df['Review'] = df['Review'].apply(lambda x: remove_punc(x))
df.head()
```

```python
#removing stopwords
sr = stopwords.words('english')
```

```python
df['Review'] = df['Review'].apply(lambda x: ' '.join(x for x in x.split() if x not in sr))
df.head()
```

```python
#removing numbers

df['Review'] = df['Review'].apply(lambda x: ' '.join(x for x in x.split() if not x.isdigit()))
df.head(10)
```

```python
#lemmatization
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])


df["Review"] = df["Review"].apply(lambda x: lemmatize_words(x))
df.head()
```

Now, our data is clean. We have removed all the stopwords, punctuation, numbers, contractions, lower cased every text, lemmatized them. The data is read for sentimental analysis. For this we will be using the TextBlob library. 

```python
sentiment=[]
sentiment_score = []
keywords = []

for x in df['Review'] :
    senti=TextBlob(x) ## sentiment analysis using TextBlob
    
    Token= [t for t in x.split()] ## string the tokens/keywords
    keywords.append(Token)

    ## conditions for division of sentiment
    if senti.sentiment.polarity > 0 and senti.sentiment.polarity <= 1 :
        sentiment_score.append(round(senti.sentiment.polarity,3))
        sentiment.append("Postive")
    elif senti.sentiment.polarity == 0 :
        sentiment_score.append(round(senti.sentiment.polarity,3))
        sentiment.append("Neutral")
    else:
        sentiment_score.append(round(senti.sentiment.polarity,3))
        sentiment.append("Negative")
```

```python
## storing the new data in our dataset
df['Keywords']=keywords
df['Sentiment']=sentiment
df['Sentiment_Score'] = sentiment_score
```
